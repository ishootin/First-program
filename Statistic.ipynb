```python
from scipy import stats
import pandas
import random

import matplotlib
import numpy
import matplotlib.pyplot as plt

#Когда-то в первом семестре мы с Гунягой проводили лаборатторную работу, на генерирование случайной величины
#В этой программе мы будем проверять насколько случайными получались наши значения

experiments_1 = [8.032, 8.027, 7.997, 8.000, 8.002,
                7.990, 7.995, 8.052, 7.995, 8.013,
                8.031, 8.025, 7.994, 7.993, 7.997,
                8.033, 8.037, 7.996, 8.041, 8.031,
                8.016, 8.016, 8.019, 8.020, 8.020,
                8.024, 8.024, 8.027, 8.026, 8.028,
                8.034, 8.034, 8.028, 8.030, 8.031,
                8.013, 8.040, 8.013, 8.014, 8.016,
                7.961, 7.970, 7.975, 7.982, 7.984]

plt.plot(experiments_1)
plt.title('Данные из лабораторной')
plt.show()

#Добавим для интереса действительно случайные величины
quantity = 50
value = 15

rgen = random.Random()

experiments_2 = []
for e in range(quantity):
    v = 0
    for r in range(value):
        v += rgen.randint(0, 1)
    experiments_2.append(v)
plt.plot(experiments_2)
plt.title('Сгенерированные случайные данные')
plt.show()

#А теперь, построим наконец распределние, и сравним
df1 = pandas.DataFrame(data={'Лабораторная': experiments_1,})
df2 = pandas.DataFrame(data={'Случайные данные': experiments_2,})

df1.plot.kde()
df2.plot.kde()
#Лабораторные данные вполне красивые, хоть и есть одно подозрительное место

#Теперь численно проверим, насколько наши данные отклоняются от "нормы"
d1 = df1['Лабораторная']
d2 = df2['Случайные данные']

print(stats.kstest(d1, 'norm', (d1.mean(), d1.std())))
print(stats.kstest(d2, 'norm', (d2.mean(), d2.std())))
```
